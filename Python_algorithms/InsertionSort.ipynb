# Insertion Sort
 
Algoritmo usado para ordenar pequeños arreglos de números

**Pseucode**

for j = 2 to A.length:
    key = A[j]
    i = j - 1
    
    while i > 0 and A[i] > key:

        A[i+1] = A[i]

        i -= 1 

    A[i + 1] = key

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import math
import time

# Insertion sort Algorithm

# Generate random arrays
def get_unsorted(size):
    
    up = random.randint(0,1000000)
    unsorted_array = np.random.randint(2,up,size = size+1)
    return unsorted_array
```

```python
# Insertion Sort algorithm

list_time = []
for iteration in range(2,1000):

    unsorted_array = get_unsorted(iteration)
    # print(unsorted_array)
    start_time = time.time()

    for i in range(1,len(unsorted_array)):
        
        key = unsorted_array[i]
        j = i - 1
        #print("Iteración", i - 1, " Key -> ",key,"J -> ",j)

        while 0 <= j and unsorted_array[j] > key:
            unsorted_array[j+1] = unsorted_array[j]
            j -= 1        
        unsorted_array[j+1] = key
    # print(unsorted_array)
    list_time.append(time.time() - start_time)

print("Fin")
time = pd.Series(list_time)
size = pd.Series(list(range(2,1000)))
```

```python
df_time = pd.DataFrame({"Time" : time,"Size":size})
print(df_time['Time'][984])
plt.hist(df_time['Time'])
plt.xlabel("Time")
plt.xlim([0,0.35])
plt.ylabel("Size")

plt.show()
```

## Insertion Sort Modification 

In this section there is a modification for the **Insertion Sort Algorithm** being as the array will be sorted in decremental orden. 

For example: 

The array [8,0,7,6,1] sorted will be [8,7,6,1,0]


```python
list_time = []
for x in range(2,1000):
    unsort = np.random.randint(0,1000,x)

    start_time = time.time()

    for i in range(1,len(unsort)):
        key = unsort[i]
        j = i - 1
        while  0 <= j and unsort[j] < key:
            unsort[j+1] = unsort[j]
            unsort[j] = key
            j -= 1

    list_time.append(time.time()-start_time)
    
```

```python
insertion_reverse = pd.DataFrame({"Time":list_time,"Size":list(range(2,1000))})
plt.scatter(insertion_reverse['Time'],insertion_reverse['Size'])
```
