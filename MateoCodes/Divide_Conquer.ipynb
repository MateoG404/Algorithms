# Insertion Sort

Es un algoritmo de ordenamiento básico en donde su lógica se basa en como ordenamos cartas nosotros, pues se supone que el primer elemento esta ordenado y luego se va ordenando el siguiente arrreglo hasta que se tenga el primer elemento en la posición adecuada y eso hasta acabar todo el array, con un tiempo de O(n²),

El pseudocodigo es :




```python
# Generar un array de 10 elementos desordenados

import numpy as np

unsorted = np.array(np.random.randint(1,10,size = 10))

print("Unsorted array ", unsorted)

```

```python
for i in range(1,len(unsorted)):
    
    key = unsorted[i] # Nuevo valor que se debe ingresar en el sub-arreglo ordenado
    j = i -1

    print(f"Clave {key}, j {j}")
    while( 0 <= j  and  key < unsorted[j]):

        unsorted[j+1] = unsorted[j]
        j -= 1
        #print(unsorted)
    unsorted[j+1] = key

print(unsorted)
```

# Ejercicio de invariante de ciclo

Implementar un algoritmo que encuentre el factorial de un número usando un while 

```python
def factorial(n):

    res = 1
    while 0 < n :
        res = res * n
        n -= 1
    return res

print(factorial(5))
```

```python
def misterio(n):

    for i in range(1,n+1):
        k = i
        print("K",k)
        while k > 1 :
            k = k/2
            print(k)

misterio(10)
```
